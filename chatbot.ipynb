```python
# 👇 Custom reader using pdfplumber to extract text from scanned/image PDFs
import pdfplumber
from llama_index.core.schema import Document

def load_pdf_with_pdfplumber(pdf_path):
    all_text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            all_text += page.extract_text() or ""
    return Document(text=all_text)

# Load your document using pdfplumber
doc = load_pdf_with_pdfplumber("docs/ncert/ncert.pdf")

# Optional: Check if the text was extracted
print(doc.text[:1000])  # Check the first 1000 characters
```

```python
from llama_index.core.node_parser import SentenceSplitter
splitter = SentenceSplitter(chunk_size=600, chunk_overlap=100)
nodes = splitter.get_nodes_from_documents([doc])
```
